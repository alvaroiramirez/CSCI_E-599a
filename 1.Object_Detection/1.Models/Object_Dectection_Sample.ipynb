```python
# looading the model
model = YOLO("yolo12m.pt")
#model = YOLO("yolo12x.pt")
```

```python
# Default example
results = model.train(data="data.yaml", epochs=300, imgsz=480,
                      device=[0, 1, 2, 3] # 4 GPus)
```

```python
# Default Example
results = model.train(data="data.yaml", epochs=300, imgsz=640,
                      device=[0, 1] # two gpus)
```

```python
results = model.train(
    data="data.yaml",
    epochs=500,         # Train longer if needed
    imgsz=1280,         # Increase image size
    device=[0, 1],      # Multi-GPU
    batch=16,   # Increase batch if VRAM allows
    cache="disk",# you can use ram if you have good ram, however is not deterministic
    augment=True,
    amp=True,           # Mixed precision training
    #multi_scale=True,   # Train with different image sizes
    conf=0.05,           # Lower confidence to improve recall
    lr0=0.001, lrf=0.0001, # Adjust learning rate
    mosaic=1.0, mixup=0.5, # Stronger augmentations
```

You can create or add more augmentations here if you did not did it on labelstudio or roboflow or whaterver software you are working on. Look at the documentation is quite simple.

```python
results = model.train(
    data="data.yaml",
    epochs=500,         # Train longer if needed
    imgsz=1280,         # Increase or decrease image size
    device=[0, 1],      # Multi-GPU
    batch=8,           # Increase batch if VRAM allows
    cache="disk",        # Load images into RAM
    amp=True,           # Mixed precision training
    #multi_scale=True,   # Train with different image sizes
    conf=0.5,
   # iou=0.45,  #

    lr0=0.002, lrf=0.00005, # Adjust learning rate
     # Stronger augmentations
)
```

```python
results = model.train(
    data="data.yaml",
    epochs=500,
    imgsz=1028,
    device=[0,1],        # Multi-GPU training
    batch=16,             # Increase batch size if possible
    augment=True,
    amp=True,     # Train on different image sizes
    optimizer="AdamW",
    conf=0.01,   # Use AdamW for better convergence
    lr0=0.0001, lrf=0.00001, # Fine-tuned learning rate
    iou=0.6,            # Use mixed precision training
    patience=50,          # Early stopping
    mosaic=1.0, mixup=0.9,
    fliplr=0.7, flipud=0.5,  # Enable flipping
    classes = [1]  # Balance classes          # Cpu loaders
)
```
